## Exploratory Data Analysis (EDA)

### Explore data day_df

```python
day_df.describe(include="all")
```

#### Pertanyaan 1

```python
filter_weekday = day_df[(day_df['weekday'] >= 1) & (day_df['weekday'] <= 6) & (day_df['holiday'] == 0)]

total_sales_weekday = filter_weekday['cnt'].sum()

average_sales_weekday = filter_weekday['cnt'].mean()

result = pd.DataFrame({
    'Total Peminjaman Sepeda (Weekday)': [total_sales_weekday],
    'Rata-rata Peminjaman Sepeda (Weekday)': [round(average_sales_weekday, 2)]
})

result
```

```python
filter_holiday = day_df[(day_df['weekday'] >= 1) & (day_df['weekday'] <= 6) & (day_df['holiday'] == 1)]

total_sales_holiday = filter_holiday['cnt'].sum()

average_sales_holiday = filter_holiday['cnt'].mean()

result = pd.DataFrame({
    'Total Peminjaman Sepeda (Holiday)': [total_sales_holiday],
    'Rata-rata Peminjaman Sepeda (Holiday)': [round(average_sales_holiday, 2)]
})

result
```

#### Pertanyaan 2

```python
total_sales_casual = day_df['casual'].sum()

average_sales_casual = day_df['casual'].mean()

result = pd.DataFrame({
    'Total Peminjaman Sepeda (Casual)': [total_sales_casual],
    'Rata-rata Peminjaman Sepeda (Casual)': [round(average_sales_casual, 2)]
})

result
```

```python
total_sales_registered = day_df['registered'].sum()

avarage_sales_registered = day_df['registered'].mean()

result = pd.DataFrame({
    'Total Peminjaman Sepeda (Registered)': [total_sales_registered],
    'Rata-rata Peminjaman Sepeda (Registered)': [round(avarage_sales_registered, 2)]
})

result
```

**Insight:**
- Kita mendapatkan informasi rata-rata peminjaman sepeda berdasarkan hari biasa (weekday) dan hari libur (holiday).
- Kita mendapatkan informasi rata-rata peminjaman sepeda berdasarkan umur yang tidak memiliki keanggotaan (casual) dan yang memiliki keanggotaan (registered).

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
data = {
    'Jenis Hari': ['Weekday', 'Holiday'],
    'Rata-rata Peminjaman': [average_sales_weekday, average_sales_holiday]
}
visualization_df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.barplot(x='Jenis Hari', y='Rata-rata Peminjaman', data=visualization_df, palette='viridis', hue='Jenis Hari', legend=True)
plt.title('Perbandingan Rata-rata Peminjaman Sepeda')
plt.ylabel('Rata-rata Peminjaman Per Hari')
plt.xlabel('Jenis Hari')
plt.ylim(0, max(visualization_df['Rata-rata Peminjaman']) * 1.1)
plt.show()
```

### Pertanyaan 2:

```python
data = {
    'Jenis Orang': ['Casual', 'Registered'],
    'Rata-rata Peminjaman': [average_sales_casual, avarage_sales_registered]
}
visualization_df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.barplot(x='Jenis Orang', y='Rata-rata Peminjaman', data=visualization_df, palette='viridis', hue='Jenis Orang', legend=True)
plt.title('Perbandingan Rata-rata Peminjaman Sepeda')
plt.ylabel('Rata-rata Peminjaman Per Hari')
plt.xlabel('Jenis Orang')
plt.ylim(0, max(visualization_df['Rata-rata Peminjaman']) * 1.1)
plt.show()
```

**Insight:**
- Kita mendapatkan gambar visualisasi dari perbandingan rata-rata peminjaman sepeda berdasarkan hari biasa (weekday) dan hari libur (holiday).
- Kita mendapatkan gambar visualisasi dari perbandingan rata-rata peminjaman sepeda berdasarkan orang yang tidak memiliki keanggotaan (casual) dan yang memiliki keanggotaan (registered).
